```python
from sklearn.datasets import load_iris
iris = load_iris()
N = 10
X = iris.data[:N, :2]

plt.plot(X.T, 'o:')
plt.xticks(range(4), [" ", " "])
plt.xlim(-0.5, 2)
plt.ylim(2.5, 6)
plt.legend([" {}".format(i+1) for i in range(N)])
plt.show()
```

```python
plt.figure(figsize=(8,8))
ax = sns.scatterplot(0, 1, data=pd.DataFrame(X), s=100, color=".2", marker="s")
for i in range(N):
    ax.text(X[i, 0] -0.05, X[i, 1] + 0.03, " {}".format(i+1))
    
plt.xlabel(" ")
plt.ylabel(" ")
plt.title("   (2 )")
plt.axis("equal")
plt.show()
```

```python
from sklearn.decomposition import PCA # PCA

pca1 = PCA(n_components=1) #pca1  1
X_low = pca1.fit_transform(X) # pca1  
X2 = pca1.inverse_transform(X_low) #     

plt.figure(figsize=(7, 7)) # 
ax = sns.scatterplot(0, 1, data=pd.DataFrame(X), s=100, color=".2", marker="s")
for i in range(N):
    d = 0.03 if X[i, 1] > X2[i, 1] else -0.04
    ax.text(X[i, 0] - 0.065, X[i, 1] + d, " {}".format(i+1))
    plt.plot([X[i, 0], X2[i, 0]], [X[i, 1], X2[i, 1]], "k--") #k-- : 
plt.plot(X2[:, 0], X2[:, 1], "o-", markersize=10)
plt.plot(X[:, 0].mean(), X[:, 1].mean(), markersize=10, marker="D")
plt.axvline(X[:, 0].mean(), c="r")
plt.axhline(X[:, 1].mean(), c="r")
plt.grid(False)
plt.xlabel(" ")
plt.ylabel(" ")
plt.title("Iris  1 ")
plt.axis("equal")
plt.show()
```

```python
pd.core.common.is_list_like = pd.api.types.is_list_like
import pandas_datareader.data as web
import datetime

symbols = [
    "SPASTT01USM661N", # US: 
    "SPASTT01JPM661N", # JP: 
    "SPASTT01EZM661N", # EZ: 
    "SPASTT01KRM661N", # KR: 
]

data = pd.DataFrame()
for sym in symbols:
    data[sym] = web.DataReader(sym, data_source='fred', 
                               start=datetime.datetime(1998, 1, 1),
                               end=datetime.datetime(2017, 12, 31))[sym]
data.columns = ["US", "JP", "EZ", "KR"]
data = data / data.iloc[0] * 100

styles = ["b-.", "g--", "c:", "r-"]
data.plot(style=styles)
plt.title("  20 ")
plt.show()
```

```python
df = ((data.pct_change() + 1).resample("A").prod() - 1).T * 100
print(df.iloc[:, :5])
```

```python
df.T.plot(style=styles)
plt.title("  20 ")
plt.xticks(df.columns)
plt.show()
```

```python
pca2 = PCA(n_components=1)
w = pca2.fit_transform(df)
```

```python
m = pca2.mean_
p1 = pca2.components_[0]
```

```python
xrange= np.linspace(1988, 2017, 20, dtype=int)
for i in np.linspace(0, 100, 5):
    plt.plot(xrange, pca2.mean_+ p1*i)
plt.plot(xrange, pca2.mean_+p1*100, label=" 100  ")
plt.plot(xrange, pca2.mean_, "ko-", lw=5, label="")
plt.title("    ")
plt.legend()
plt.show()
```

```python
from scipy.misc import derivative

print(derivative(f, 0, dx=1e-6))
print(derivative(f, 1, dx=1e-6))
```

```python
def relu(x):
    return np.where(x>0, x, 0)

xx = np.linspace(-2, 2, 100)
plt.plot(xx, relu(xx))
plt.plot(0, 0, marker="o", ms=10)
plt.title("ReLu")
plt.xlabel("$x$")
plt.ylabel("ReLU(x)")
plt.show()
```

```python
#네이버 스토어  상품정보 크롤링
url = "https://shopping.naver.com/v1/products?_nc_=1582642800000&subVertical=PET&page=1&pageSize=100&sort=RECENT&displayType=CATEGORY_HOME&includeZzim=true&includeViewCount=false&includeStoreCardInfo=false&includeStockQuantity=false&includeBrandInfo=false&includeBrandLogoImage=false&includeRepresentativeReview=false&includeListCardAttribute=false&includeRanking=false&includeRankingByMenus=false&includeStoreCategoryName=false&storeId=100150961&standardSizeKeys=&standardColorKeys=&optionFilters=&attributeValueIds=&attributeValueIdsAll=&certifications="
response = requests.get(url)
result = response.json()
product = result["products"]
prdinfo = []

for i in range(len(product)):
    idx_product = product[i]
    tags = []
    for j in range(len(idx_product["tags"])):
        tags.append(idx_product["tags"][j]["name"])
        tag_str = ""
        for a in tags:
            tag_str = tag_str + a +","
    
    prdinfo.append({'상품정보' : idx_product["name"], 
                    '가 격' : idx_product["mobileDiscountPrice"], 
                    '정보' :idx_product["contentText"],
                    '태그': tag_str})
    
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
sy = pd.Series(iris.target, dtype="category")
sy = sy.cat.rename_categories(iris.target_names)
df['species'] = sy
```

```python
df
```

```python
df[df.species=="setosa"]["petal length (cm)"].hist() #setosa 종의 꽃잎길이 분포 : 단봉분포
plt.show()
```

```python
df["petal length (cm)"].hist() # 붓꽃의 꽃잎의 길이 분포 : 다봉분포
plt.show()
```

```python
np.random.seed(1)
x = np.random.normal(size=1000)
x = np.hstack([x, 5*np.ones(50)])
bins = np.linspace(-6, 6, 12*4+1)
ns, _ = np.histogram(x, bins=bins)

sample_mean = np.mean(x)
smaple_median = np.median(x)
mode_index = np.argmax(ns)
sample_mode = 0.5*(bins[mode_index]+ bins[mode_index+1])

sns.distplot(x, bins=bins)
plt.axvline(sample_mean, c='k', ls=":", label="표본평균")
plt.axvline(sample_median, c="y", ls="--", label="표본중앙값")
plt.axvline(sample_mode, c='g', ls="-", label="표본최빈값")
plt.title("표본평균, 표본중앙값, 표본최빈값의 차이")
plt.xlabel("x")
plt.legend()
plt.show()


                   
                   
```
